## Snapshot shufflings - instant events ##
We here present randomization techniques which shuffle events in snapshots. All functions take a `snapshot_sequence()` object as input, and return a `snapshot_sequence()` object.

### Permutation of events in snapshots $\text{P}[\mathbf{t}]$: `P__t(seq_data)` ###
This function randomly permutates links within each snapshot of a `snapshot_sequence()` object `seq_data`.

```python
seq_data = tnu.read_snapshot_sequence("./data/timeset.dat")
new_seq = tnr.P__t(seq_data)
```

```python
list_lk_0 = seq_data.data[1503].display()
list_lk_1 = new_seq.data[1503].display()

fig = plt.figure(figsize=(8,4))
ax_0 = fig.add_axes([0,0,0.5,1])
ax_1 = fig.add_axes([0.5,0,0.5,1])
import networkx as nx
G = nx.Graph()
G.add_edges_from(list_lk_0)
nx.draw_networkx(G,ax=ax_0)
G = nx.Graph()
G.add_edges_from(list_lk_1)
nx.draw_networkx(G,ax=ax_1)
ax_0.set_xticks([])
ax_0.set_yticks([])
ax_1.set_xticks([])
ax_1.set_yticks([])
```

### Permutation of events in snapshots with preserved degrees $\text{P}[\mathbf{d}]$: `P__d(seq_data,link_threshold=20,n_iter=5)` ###
This function randomly permutates links within each snapshot while preserving the instantaneous degrees of the nodes, for a `snapshot_sequence()` object `seq_data`. We use to algorithms to do so. If the number of links in the snapshot if larger than `link_threshold` then we use a configuration model. If it is lower, we use the Sneppen-Maslov algorith, which we iterate $n_\text{iter}\times n_L$ times.

This choice is due to the fact that exploring the set of possible networks via the configuration model can be extremely inefficient for small networks.

```python
seq_data = tnu.read_snapshot_sequence("./data/timeset.dat")
seq_data.data = {t:seq_data.data[t] for t in seq_data.data if t < 1510}
new_seq = tnr.P__d(seq_data)
```

```python
list_lk_0 = seq_data.data[1503].display()
list_lk_1 = new_seq.data[1503].display()

fig = plt.figure(figsize=(8,4))
ax_0 = fig.add_axes([0,0,0.5,1])
ax_1 = fig.add_axes([0.5,0,0.5,1])
import networkx as nx
G = nx.Graph()
G.add_edges_from(list_lk_0)
nx.draw_networkx(G,ax=ax_0)
G = nx.Graph()
G.add_edges_from(list_lk_1)
nx.draw_networkx(G,ax=ax_1)
ax_0.set_xticks([])
ax_0.set_yticks([])
ax_1.set_xticks([])
ax_1.set_yticks([])
```

## Snapshot shufflings - events ##
We here present a randomization technique which shuffles events globally.

### Global contact shuffling $\text{P}[p(\mathbf{t},\tau)]$: `P__pttau(lks_data,dt)` ###
This function shuffles contacts between all possible links of a `link_timeline()` object `lks_data` but preserves their starting time and duration. The time step duration `dt` is required.

```python
lks_data = tnu.read_link_timeline("./data/lks.dat")
new_lks = tnr.P__pttau(lks_data,1)
```

Test of the preservation of the activity timeline: 

```python
TL0 = tnm.activity_timeline(tnu.link_timeline_to_snapshot_sequence(lks_data,1,0,60480))
TL1 = tnm.activity_timeline(tnu.link_timeline_to_snapshot_sequence(new_lks,1,0,60480))

tmin = 1000
tmax = 2000

fig = plt.figure(figsize=(8,2))
x = range(len(TL0))[tmin:tmax]
plt.plot(x,TL0[tmin:tmax],'k-')
plt.plot(x,TL1[tmin:tmax],'r-')
plt.show()
```

Test of the modification of the weights distribution:

```python
list_w0 = list(tnm.weights(lks_data).values())
list_w1 = list(tnm.weights(new_lks).values())

from math import log

fig = plt.figure(figsize=(4,4))
wmax = max(list_w0)
x = 2.**(np.arange(0,log(wmax)+1,0.5))
h0 = np.histogram(list_w0,bins=x)
h1 = np.histogram(list_w1,bins=x)
plt.loglog(x[1:],h0[0],'ko')
plt.loglog(x[1:],h1[0],'r^')
plt.show()
```
